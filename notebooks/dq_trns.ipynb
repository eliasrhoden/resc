U\\
V\\
W
\end{bmatrix}
$$


```python
import numpy as np


cos120 = -0.5
sin120 = np.sqrt(3)/2

M = np.array([[1,cos120,cos120],
             [0, sin120, -sin120],
             [1,1,1]])

M

```

```python
from numpy import linalg

M_inv = linalg.inv(M)
M_inv
```

```python
import matplotlib.pyplot as plt

def plot_phase(Ud,Uq,alpha):
    
    X = np.array([Ud,Uq,1])
    Y = M_inv @ X
    print(Y)
    
    plt.axis('equal')
    plt.plot(Ud,0,'*',label='Ud')
    plt.plot(0,Uq,'*',label='Uq')
    
    plt.plot([0,Y[0]],[0,0],label='U')
    plt.plot([0,-0.5*Y[1]],[0,sin120*Y[1]],label='V')
    plt.plot([0,-0.5*Y[2]],[0,-sin120*Y[2]],label='W')
    plt.legend()
    
    
    
    
plot_phase(0,1,0)
    
```

```python
plot_phase(0,-1,0)
```

```python
plot_phase(1,0,0)
```

```python
plot_phase(0,0.2,0)
```
